```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree")
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
rating = soup.find('div',class_="_3LWZlK")
rating
```

```python
rating.text
```

```python
rating = soup.find_all('div',class_="_3LWZlK")
name = soup.find_all('div',class_="_4rR01T")
review = soup.find_all('div',class_="t-ZTKy")
```

```python
review
```

```python
laptop_rating=[]
for i in rating:
    laptop_rating.append(i.text)

laptop_name=[]
for i in name:
    laptop_name.append(i.text)

laptop_review=[]
for i in review:
    laptop_review.append(i.text)
```

```python
import pandas as pd
```

```python
laptop=pd.DataFrame({})
laptop['laptop_name']=laptop_name
laptop['laptop_rating']=laptop_rating
laptop['laptop_review']=laptop_review
```
