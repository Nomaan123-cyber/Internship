```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://bookpage.com/')
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
name=soup.find_all('h4',class_='italic')
name
```

```python
Book_name=[]
for i in name:
    Book_name.append(i.text.replace('\n',''))
Book_name
```

```python
aname=soup.find_all('p',class_='sans bold')
aname
```

```python
Author_name=[]
for i in aname:
    Author_name.append(i.text.replace('\n',''))
Author_name
```

```python
genre=soup.find_all('p',class_='genre-links hidden-phone')
genre
```

```python
Genre=[]
for i in genre:
    Genre.append(i.text.replace('\n',''))
Genre
```

```python
review=soup.find_all('div',class_='read-full')
review
```

```python
Review=[]
for i in review:
    Review.append(i.text.replace('\n',''))
Review
```

```python
import pandas as pd
```

```python
Book=pd.DataFrame()
Book['Author_name']=Author_name
Book['Genre']=Genre
Book['Review']=Review
Book
```
